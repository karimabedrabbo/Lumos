# Crop Original and  Ground Truth Images to Disc 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import scipy
from PIL import Image
import glob
```

```python
def create_contours(thresh_disc, original, path_disc, path_orig, i):

    
    _, cup_contours, _ = cv2.findContours(thresh_disc, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cup_contours = sorted(cup_contours, key = cv2.contourArea, reverse = True)[:5]
    # Scan the cup contours for largest item

    array_sizes = [len(numpoints) for numpoints in cup_contours]

    largest = max(array_sizes)
    index_of_object = array_sizes.index(largest)

    # Largest detected contour
    cup_cnt = cup_contours[index_of_object] 


    # Find the largest and smallest y value
    y_values = [points[0][1] for points in cup_cnt]

    crop_constant = 0
    largest_y = max(y_values)
    smallest_y = min(y_values)

    vertical_disk = largest_y - smallest_y


    # Find the largest and smallest x value
    x_values = [points[0][0] for points in cup_cnt]

    largest_x = max(x_values)
    smallest_x = min(x_values)

    horizontal_disk = largest_x - smallest_x

    max_distance = max(vertical_disk, horizontal_disk)




    # display the final image
    # cv2.drawContours(cup_threshed, [cup_cnt], -1, (128,255,0), 3)
    # final = Image.fromarray(cup_threshed).show()

    image_constant = max_distance + 50
    print(image_constant)

    # we want disc height and width to be that specific length, so basically:
        # min_x = largest_x - smallest_x



    mdpt_vert = largest_y - (vertical_disk/2)
    mdpt_horiz = largest_x - (horizontal_disk/2)
    print(smallest_x,largest_x,smallest_y,largest_y)

    new_smallest_y = int(smallest_y - ((image_constant/2) - (mdpt_vert - smallest_y)))
    new_largest_y = int(largest_y + ((image_constant/2) - (largest_y - mdpt_vert)))
    new_smallest_x = int(smallest_x - ((image_constant/2) - (mdpt_horiz - smallest_x)))
    new_largest_x = int(largest_x + ((image_constant/2) - (largest_x - mdpt_horiz)))
    print(new_smallest_x,new_largest_x,new_smallest_y,new_largest_y)
    
    if new_smallest_y <= 0:
        new_smallest_y = 0
    elif new_largest_y <= 0:
        new_largest_y = 0
    elif new_smallest_x <= 0:
        new_smallest_x = 0
    elif new_largest_x <= 0:
         new_largest_x = 0
    
    new_thresh_disc = thresh_disc[new_smallest_y:new_largest_y, new_smallest_x:new_largest_x]
    bw = Image.fromarray(new_thresh_disc)
    bw.save(path_disc+str(i)+ ".png","PNG")
    
#     new_thresh_cup = thresh_cup[new_smallest_y:new_largest_y, new_smallest_x:new_largest_x]
#     cv2.imwrite(path_cup+str(i)+".png",new_thresh_cup)
    
    
    roi = original[new_smallest_y:new_largest_y, new_smallest_x:new_largest_x]
    color = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)
#     plt.imshow(color)
    #cv2.imwrite(path_orig+str(i)+".png",color)
    color = Image.fromarray(color)
    color.save(path_orig+str(i)+ ".png","PNG")
```

# Running on DRISHTI Testing Set

```python
glauc_orig = glob.glob1("glaucoma","*.jpg")
not_glauc_orig = glob.glob1("not_glaucoma","*.jpg")

glauc_thresh_disc = glob.glob1("glaucoma_truths","*.jpg")
not_glauc_thresh_disc = glob.glob1("not_glaucoma_truths","*.jpg")

glauc_orig.sort()
glauc_thresh_disc.sort()
not_glauc_orig.sort()
not_glauc_thresh_disc.sort()

# print(glauc_orig)
# print("\n\n")
# print(glauc_thresh_disc)



for i in range(len(glauc_orig)):
    original = glauc_orig[i]
    thresh_disc = glauc_thresh_disc[i]

    original = cv2.imread("glaucoma/"+original)
    thresh_disc = cv2.imread("glaucoma_truths/"+thresh_disc)
    thresh_disc = cv2.cvtColor(thresh_disc, cv2.COLOR_BGR2GRAY)
    ret, thresh_disc = cv2.threshold(thresh_disc, 50, 255, 0)
    create_contours(thresh_disc, original, "cropped_glauc_truths/", "cropped_glaucoma/", glauc_orig[i])
    
    
# print(not_glauc_orig)
# print("\n\n")
# print(not_glauc_thresh_disc)


for i in range(len(not_glauc_orig)):
    original = not_glauc_orig[i]
    thresh_disc = not_glauc_thresh_disc[i]

    original = cv2.imread("not_glaucoma/"+original)
    thresh_disc = cv2.imread("not_glaucoma_truths/"+thresh_disc)
    thresh_disc = cv2.cvtColor(thresh_disc, cv2.COLOR_BGR2GRAY)
    ret, thresh_disc = cv2.threshold(thresh_disc, 50, 255, 0)
    create_contours(thresh_disc, original, "cropped_not_glauc_truths/", "cropped_not_glaucoma/", not_glauc_orig[i])

   

# original = cv2.imread("glaucoma/239.jpg")
# thresh = cv2.imread("glaucoma_truths/239.jpg")
# thresh = cv2.cvtColor(thresh, cv2.COLOR_BGR2GRAY)
# ret, thresh = cv2.threshold(thresh, 50, 255, 0)
# create_contours(thresh, original, "cropped_glauc_truths/", "cropped_glaucoma/", 239)

```
