```python
import cv2
import numpy as np
import sys
import os
import glob
from keras.preprocessing import image
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import scipy
from scipy import misc
import scipy.io
from keras.preprocessing import image
```

```python
def sort(path, im_type, backend):

    images = glob.glob1(path,"*." + str(im_type))
    im_type_length = len(im_type) + 1
    images = [element[:-im_type_length] for element in images]
    images.sort(key=int)
    images = [element + "." + str(im_type) for element in images]

    if backend == "keras" or backend == "Keras" or backend == "keras":
        images = [np.array(image.load_img(path + "/" + fname)) for fname in images]
    if backend == "opencv" or backend == "openCV" or backend == "OPENCV" or backend == "cv2":
        images = [np.array(cv2.imread(path + "/" + fname)) for fname in images]
    if backend == "PIL" or backend == "pil":
        images = [np.array(Image.open(path + "/" + fname)) for fname in images]
    return images
```

```python
disc_path = "/Users/karimabedrabbo/Desktop/work2/cropped_glauc_discs_png"
cup_path = "/Users/karimabedrabbo/Desktop/work2/cropped_glauc_cups_png"
vessels_path = "/Users/karimabedrabbo/Desktop/work2/cropped_glauc_vessels_png"
output_path = "/Users/karimabedrabbo/Desktop/work2/cropped_glauc_masks_png"

disc_gt_imgs = sort(disc_path, im_type="png", backend="keras")
cup_gt_imgs = sort(cup_path, im_type="png", backend="keras")
vessels_gt_imgs = sort(vessels_path, im_type="png", backend="keras")
```

```python
def create_masks(discs, cups, vessels):
    total = len(discs)
    final = []
    for i in range(total):
        print(i)
        disc = np.asarray(discs[i])
        cup = np.asarray(cups[i])
        vessel = np.asarray(vessels[i])
        cup_gray = cv2.cvtColor(cup, cv2.COLOR_BGR2GRAY)
        disc_gray = cv2.cvtColor(disc, cv2.COLOR_BGR2GRAY)
        vessels_gray = cv2.cvtColor(vessel, cv2.COLOR_BGR2GRAY)
        start = np.zeros([disc.shape[0],disc.shape[1],3],dtype=np.uint8)
        start[disc_gray > 247] = [255, 0, 0]
        start[cup_gray > 247] = [0, 255, 0]
        start[vessels_gray > 247] = [0, 0, 255]
        start_gray = cv2.cvtColor(start, cv2.COLOR_BGR2GRAY)
        start[start_gray < 7] = [255, 255, 0]
        final.append(start)
    return final
```

```python
final = create_masks(disc_gt_imgs, cup_gt_imgs, vessels_gt_imgs)
```

```python
def save_images(aug_path, aug_imgs):
    for i in range(len(aug_imgs)):
        aug = misc.toimage(aug_imgs[i])
        aug.save(aug_path + "/" + str(i + 1) + ".png")
```

```python
save_images(output_path, final)
```
